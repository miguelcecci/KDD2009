```python
import pandas as pd
import tensorflow as tf
```

```python
validation = pd.read_pickle('../validation_set.pkl')
```

```python
validation
```

```python
from ast import literal_eval
def getData(index):
    return literal_eval(validation['data'][index])
```

```python
with tf.Session() as sess:
    saver = tf.train.import_meta_graph('../saved_models/saved_models/kdd_model_3.ckpt.meta')
    saver.restore(sess, tf.train.latest_checkpoint('../saved_models/saved_models/'))
    graph = tf.get_default_graph()
#     print(graph.get_operations())
    x = graph.get_tensor_by_name("Placeholder:0")
    for i in range(len(validation)):
        feed = {x:[getData(i)]}
        restore_op = tf.nn.tanh(graph.get_tensor_by_name('Tanh:0'))
        a = sess.run(restore_op, feed)
        
        
        
        print('id: {} - output: {} - label: {}'.format(i, a, validation['label'][i]))
```
