Como o numero de valores -1 é muito superior, na hora de gerar o batch o modelo vai tender a gerar outputs negativos, por conta disso, farei uma condição para equalizar o carregamento de batches com valores -1 iguais ao de 1. Vai deixar mais lento na hora do treino, mas deve funcionar.

```python
import random
def loadbatch(dataframe, batch_size, objective):
    if objective == 'a':
        aux = 0
    elif objective == 'u':
        aux = 1
    elif objective == 'c':
        aux = 2
    else:
        print('objetivo inexistente')
    batch_x = []
    batch_y = []
    for i in range(int(batch_size/2)):
        randval = random.randrange(len(dataframe))
        while literal_eval(dataframe['label'][randval])[aux]<0:
            randval = random.randrange(len(dataframe))
        batch_x.append(literal_eval(dataframe['data'][randval]))
        batch_y.append([literal_eval(dataframe['label'][randval])[aux]])
    for i in range(int(batch_size/2)):
        randval = random.randrange(len(dataframe))
        batch_x.append(literal_eval(dataframe['data'][randval]))
        batch_y.append([literal_eval(dataframe['label'][randval])[aux]])
    return batch_x, batch_y
```

```python
import tensorflow as tf
from tensorflow.contrib import slim
```

```python
x = tf.placeholder(tf.float32, shape=[None, 7890], name='input_placeholder')
y_ = tf.placeholder(tf.float32, shape=[None, 1])
learning_rate = tf.placeholder('float', [])
```

```python
def model(hl):
#     hl = slim.fully_connected(hl, 8100)
    hl = slim.fully_connected(hl, 2000)
    print(hl.shape)
#     hl = slim.fully_connected(hl, 2000)
#     print(hl.shape)
#     hl = tf.reshape(hl, [-1, 90, 90, 1])
#     print(hl.shape)
#     hl = tf.layers.batch_normalization(slim.conv2d(hl, 16, [7, 7]))
#     print(hl.shape)
#     hl = slim.max_pool2d(hl, [3,3], stride=2)
#     residual = hl
#     print(hl.shape)
#     hl = tf.layers.batch_normalization(slim.conv2d(hl, 32, [5, 5]))
#     print(hl.shape)
#     hl = slim.max_pool2d(hl, [3,3], stride=2)
#     print(hl.shape)
#     hl = tf.layers.batch_normalization(slim.conv2d(hl, 32, [3, 3]))
#     print(hl.shape)
#     hl = slim.max_pool2d(hl, [3,3], stride=2)
    print(hl.shape)
#     hl = slim.flatten(hl)
    print(hl.shape)
#     hl = tf.layers.batch_normalization(slim.fully_connected(hl, 1000))
    print(hl.shape)
#     hl = tf.layers.batch_normalization(slim.fully_connected(hl, 100))
    print(hl.shape)
    hl = tf.layers.batch_normalization(slim.fully_connected(hl, 1))
    print(hl.shape)
    return(tf.sigmoid(hl))
```

```python
model_1 = model(x)
```

```python
loss = tf.reduce_mean((model_1-y_)**2)
train = tf.train.AdadeltaOptimizer(learning_rate).minimize(loss)
```

```python
init = tf.global_variables_initializer()
saver = tf.train.Saver()
```

```python
batch_size = 6
steps = 5000
rate = 1
```

```python
with tf.Session() as sess:
    sess.run(init)
    for i in range(steps):
        batch_x, batch_y = loadbatch(data, batch_size, objective='a')
        feed = {x:batch_x, y_:batch_y, learning_rate:rate}
        sess.run(train, feed_dict=feed)
        
        if i%100 == 0:
            val_x, val_y = loadbatch(validation, 100, objective='a')
            feed_validation = {x:val_x, y_:val_y}
            print(sess.run(model_1, feed_dict=feed_validation))
            print(sess.run(loss, feed_dict=feed_validation))
            print(val_y)
            if i>1000:
                rate = 3e-4
            
    location = './saved_models/kdd_model_3.ckpt'
    save_path = (saver.save(sess, location))
    print('model saved in', location)
```
