Como o numero de valores -1 é muito superior, na hora de gerar o batch o modelo vai tender a gerar outputs negativos, por conta disso, farei uma condição para equalizar o carregamento de batches com valores -1 iguais ao de 1. Vai deixar mais lento na hora do treino, mas deve funcionar. 

```python
import random
def loadbatch(dataframe, batch_size, objective):
    if objective == 'a':
        aux = 0
    elif objective == 'u':
        aux = 1
    elif objective == 'c':
        aux = 2
    else:
        print('objetivo inexistente')
    batch_x = []
    batch_y = []
    for i in range(int(batch_size)):
        randval = random.randrange(len(dataframe))
        if random.random() > 0.5:
            while literal_eval(dataframe['label'][randval])[aux]<0:
                randval = random.randrange(len(dataframe))
        else:
            while literal_eval(dataframe['label'][randval])[aux]>0:
                randval = random.randrange(len(dataframe))
        batch_x.append(literal_eval(dataframe['data'][randval]))
        if literal_eval(dataframe['label'][randval])[aux] < 0:
            batch_y.append([0])
        else:
            batch_y.append([literal_eval(dataframe['label'][randval])[aux]])
    return batch_x, batch_y
```

```python
model_1 = (model(x))
```

```python
# loss = tf.reduce_mean((y_ - model_1)**2)
loss = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(labels=y_, logits=model_1))
train = tf.train.AdamOptimizer(learning_rate).minimize(loss)
```

```python
init = tf.global_variables_initializer()
saver = tf.train.Saver()
```

```python
batch_size = 40
steps = 5000
rate = 0.0001
```

```python
with tf.Session() as sess:
    sess.run(init)
    for i in range(steps):
        batch_x, batch_y = loadbatch(data, batch_size, objective='a')
        feed = {x:batch_x, y_:batch_y, learning_rate:rate}
        sess.run(train, feed_dict=feed)
        
        if i%10 == 0:
            val_x, val_y = loadbatch(validation, 10, objective='a')
            feed_validation = {x:val_x, y_:val_y}
            print(sess.run(model_1, feed_dict=feed_validation))
            print(sess.run(loss, feed_dict=feed_validation))
            print(val_y)
            if i>1000:
                rate = 3e-4
            
    location = './saved_models/kdd_model_3.ckpt'
    save_path = (saver.save(sess, location))
    print('model saved in', location)
```
