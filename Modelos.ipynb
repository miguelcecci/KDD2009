```python
import random
def loadbatch(dataframe, batch_size):
    batch_x = []
    batch_y = []
    for i in range(batch_size):
        randval = random.randrange(len(dataframe))
        batch_x.append(literal_eval(dataframe['data'][i]))
        batch_y.append(literal_eval(dataframe['label'][i]))
    return batch_x, batch_y
```

```python
import tensorflow as tf
from tensorflow.contrib import slim
```

```python
x = tf.placeholder(tf.float32, shape=[None, 7890])
y_ = tf.placeholder(tf.float32, shape=[None, 3])
```

```python
def model(hl):
#     hl = slim.fully_connected(hl, 8100)
    hl = slim.fully_connected(hl, 2000)
    print(hl.shape)
#     hl = slim.fully_connected(hl, 2000)
#     print(hl.shape)
#     hl = tf.reshape(hl, [-1, 90, 90, 1])
#     print(hl.shape)
#     hl = tf.layers.batch_normalization(slim.conv2d(hl, 16, [7, 7]))
#     print(hl.shape)
#     hl = slim.max_pool2d(hl, [3,3], stride=2)
#     residual = hl
#     print(hl.shape)
#     hl = tf.layers.batch_normalization(slim.conv2d(hl, 32, [5, 5]))
#     print(hl.shape)
#     hl = slim.max_pool2d(hl, [3,3], stride=2)
#     print(hl.shape)
#     hl = tf.layers.batch_normalization(slim.conv2d(hl, 32, [3, 3]))
#     print(hl.shape)
#     hl = slim.max_pool2d(hl, [3,3], stride=2)
#     print(hl.shape)
#     hl = slim.flatten(hl)
#     print(hl.shape)
    hl = tf.layers.batch_normalization(slim.fully_connected(hl, 500))
    print(hl.shape)
    hl = tf.layers.batch_normalization(slim.fully_connected(hl, 100))
    print(hl.shape)
    hl = tf.layers.batch_normalization(slim.fully_connected(hl, 3))
    print(hl.shape)
    return(tf.tanh(hl))
```

```python
model_1 = model(x)
```

```python
loss = tf.reduce_mean((model_1-y_)**2)
train = tf.train.AdadeltaOptimizer(3e-1).minimize(loss)
```

```python
init = tf.global_variables_initializer()
saver = tf.train.Saver()
```

```python
batch_size = 20
steps = 6000
```

```python
with tf.Session() as sess:
    sess.run(init)
    for i in range(steps):
        batch_x, batch_y = loadbatch(data, batch_size)
        feed = {x:batch_x, y_:batch_y}
        sess.run(train, feed_dict=feed)
        
        if i%100 == 0:
            val_x, val_y = loadbatch(validation, 100)
            feed_validation = {x:val_x, y_:val_y}
            print(sess.run(loss, feed_dict=feed_validation))
            
    location = './saved_models/kdd_model_1.ckpt'
    save_path = (saver.save(sess, location))
    print('model saved in', location)
```
